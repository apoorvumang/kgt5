```python
def addNameThing(query):
    query = 'SPARQL\n' + query
    query = query.replace('SELECT DISTINCT ?x', 'SELECT DISTINCT ?name')
    x = query.find('FILTER')
    query = query[:x] + 'FILTER (isLiteral(?name) AND langMatches(lang(?name), \'en\'))\n' + query[x:]
    x = query.find('}\n') - 1
    query = query[:x]+ '\n?x ns:type.object.name ?name .'  + query[x:]
    query += ';'
    return query

def addNameThing2(query, qid):
    query = 'SPARQL\n' + query
    query = query.replace('SELECT DISTINCT ?x', 'SELECT DISTINCT ?name, \'{}\''.format(qid))
    x = query.find('FILTER')
    query = query[:x] + 'FILTER (isLiteral(?name) AND langMatches(lang(?name), \'en\'))\n' + query[x:]
    x = query.find('}\n') - 1
    query = query[:x]+ '\n?x ns:type.object.name ?name .'  + query[x:]
    query += ';'
    return query
```
